```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 4, 9, 16, 20]

plt.plot(x,y)
plt.show() # show를 꼭 같이 써야함
```

```python
x = np.arange(-10,11,1)     # 정의역 선언
y = x**2
y1 = 3*x + 50
y2 = -x + 50

plt.Figure(figsize=(10,5))
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x, (lambda n:3*n**2)(x))
plt.show()
```

```python
plt.plot(x,y,x,y1,x,y2,x,(lambda n:3*n**2)(x))
plt.show()
```

```python
plt.figure(2,figsize=(10,5))
plt.plot(x,y)
plt.figure(1,figsize=(10,5))
plt.plot(x,y1)
plt.figure(1)
plt.plot(x,y2)
plt.figure(2)
plt.plot(x,(lambda n:3*n**2)(x))

plt.show()
```

```python
plt.subplot(2,2,1)          # axes 지정
plt.plot(x,y)
plt.title('axes 1 (2,2)')

plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title('axes 2 (2,2)')

plt.subplot(2,2,3)
plt.plot(x,y2)
plt.title('axes 3 (2,2)')

plt.subplot(2,2,4)
plt.plot(x,(lambda n:3*n**2)(x))
plt.title('axes 4 (2,2)')

plt.tight_layout()

plt.show()
```

```python
plt.plot(x, y, x, (lambda n:3*n**2)(x))
plt.xlim(-3,3)
plt.ylim(-50,50)

plt.show()
```

```python
plt.rc('font', family='Gulim')

plt.plot(x, y, 'o--b', x, y1, '+-r', x, y2, 's:k', linewidth=2, markersize=7)
plt.title('제목') # 기본인 labtool 폰트는 영어 위주라 네모로 출력됨
plt.xlabel('x 축')
plt.ylabel('y 축')
plt.ylim(-10, 130)
plt.grid(True)
plt.legend(['범례1', '범례2', '범례3'], loc='upper right') # upper center bottom / left center right

plt.show()
```

```python
df = pd.DataFrame({'year':[n for n in range(2000,2010)], 
                   'sales': np.random.randint(2000,10000,size=10)})
df
```

```python
df.plot(x='year', y='sales',style='g--o',title='연도별 매출액')
plt.show()
```

```python
df[(df['year']>=2003) & (df['year']<=2007)].\
    plot(x='year', y='sales', style='c-o', title='연도별 매출액') # df['year']과 df.loc[:,'year']는 동등
plt.xlabel('연도')
plt.ylabel('매출액')
plt.xticks(range(2003,2008))
plt.show()
```

```python
df[(df['year']>=2003) & (df['year']<=2007)].plot(x='year', y='sales', kind='bar', title='연도별 매출액')
plt.xlabel('연도')
plt.ylabel('매출액')
plt.show()
```

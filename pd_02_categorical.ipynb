```python
df_class, df_class1 = df[['class']].copy(), df[['class']] # 원본 변형 막기 위해 copy
df_class.head()
```

```python
# label encoding
codes, uniques = pd.factorize(df_class['class'])
codes, uniques
```

```python
df_class['class_label'] = codes
df_class.head()
```

```python
df_class['class_label2'] = df_class['class'].map({'First':1, 'Second':2, 'Third':3})
df_class.head()
```

```python
# one hot encoding
df_class2 = pd.get_dummies(df_class1, columns = ['class'])
df_class2
```

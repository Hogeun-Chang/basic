```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# df = sns.load_dataset("tips") # 온라인 접근용 github
df = pd.read_csv('datasets/seaborn-data/tips.csv')
df.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=df, style='smoker')
plt.show() # smoker 여부에 따라서 데이터 분산이 다르게 나타남. 일반화를 위해서 smoker 제거할 것인가 고민해보는 상황.
```

```python
sns.scatterplot(x='total_bill', y='tip', size='size', sizes=(15,100), data=df)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=df, ci=None)  # lmplot > 회귀선 with 신뢰구간, ci=None 으로 신뢰구간 제거
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', col='time', row='smoker', data=df)
plt.show()
```

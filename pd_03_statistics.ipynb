```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasets/성적통계.csv', encoding='ansi')  # excel 저장시 UTF-8로 저장하지 않으면, MS949로 저장됨. MS949는 ansi 호환되기에 encoding = ansi 있어야 읽어올 수 있음, 없으면 error 발생
df.head()
```

```python
df.drop(['학년','반','번호','성별'], axis=1, inplace=True)    # 학년, 반, 번호, 성별은 통계에 영향 미치지 않으므로 제거
df.head()
```

```python
df.set_index('성명', inplace=True) # 성명을 행 index로 지정
df.head()
```

```python
df.fillna(0, inplace=True)
df
```

```python
df = df.astype(pd.Int32Dtype())
df
```

```python
print(df.count(),df.count(axis=1)) # 행방향 count, 열방향 count
```

```python
print(df.mean(), df.mean(axis=1))   # 행방향 평균, 열방향 평균
```

```python
print(df.var()) # 분산
```

```python
print(df.std()) # 표준편차
```

```python
print(df.min(),df.max())    # 최소, 최대
```

```python
df.mode()
```

```python
df.median()
```

```python
df['국어'].value_counts()
```

```python
df.corr()
```
